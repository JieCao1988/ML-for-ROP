```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import AdaBoostRegressor
from sklearn.ensemble import RandomForestRegressor# Instantiate model with 1000 decision trees
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
input_a = np.random.rand(1000)
input_b = np.random.rand(1000)
input_c = np.random.rand(1000)


def my_func(data):
    a = data[0]
    b = data[1]
    c = data[2]
    return a + b**4 - 0.5*np.sin(c/10) + np.random.rand()*0.1
    #return np.sin(a) + np.sin(b) + np.sin(c)
```

```python
samples = [input_a, input_b, input_c]
samples = np.rot90(samples)
print(samples)
```

```python
results = np.zeros(len(samples))
```

```python
for i in range(len(samples)):
    results[i] = my_func(samples[i])
```

```python
train_x, test_x, train_y,  test_y = train_test_split(samples,results, test_size=0.5, random_state=42)
```

```python
def evaluate_model(rf, train_x, test_x, train_y,  test_y):
    rf.fit(train_x, train_y)
    predictions = rf.predict(test_x)

    plt.figure(figsize=(5,5))

    # Use the forest's predict method on the test data
    # Calculate the absolute errors
    errors = np.abs(predictions - test_y)# Print out the mean absolute error (mae)
    print('Mean Absolute Error:', round(np.mean(errors), 10), 'units')
    print(f'R2 = {r2_score(test_y,predictions)}')
    plt.scatter(test_y,predictions, s=2, c="black",label="Data sample", alpha=1)
    plt.plot([0,np.max([test_y,predictions])],[0,np.max([test_y,predictions])],c="red", linewidth=1, linestyle="-.", label="perfect prediction")
    plt.xlabel("Truth")
    plt.ylabel("Model")
    plt.legend()
    plt.grid()
    plt.text(0.061, 0.001, f'$R^2$ = {np.round(r2_score(test_y,predictions),3)}')
    
    return r2_score(test_y,predictions)
```

```python
rf = RandomForestRegressor(n_estimators = 30, random_state = 42, max_depth=5)

evaluate_model(rf, train_x, test_x, train_y,  test_y)
```

```python
rf = MLPRegressor()

#check documentation for all the options https://scikit-learn.org/stable/modules/generated/sklearn.neural_network.MLPRegressor.html

evaluate_model(rf, train_x, test_x, train_y,  test_y)
```

```python
rf = AdaBoostRegressor()

evaluate_model(rf, train_x, test_x, train_y,  test_y)
```

```python
rf = KNeighborsRegressor()

evaluate_model(rf, train_x, test_x, train_y,  test_y)
```

```python
rf = LinearRegression()
evaluate_model(rf, train_x, test_x, train_y,  test_y)
```

```python
def evaluate_model_fast(rf, train_x, test_x, train_y,  test_y):
    rf.fit(train_x, train_y)
    predictions = rf.predict(test_x)
    return 1/(1-r2_score(test_y,predictions))


hyper_1 = np.arange(1,200,1)
hyper_out = np.zeros(len(hyper_1))


for i in range(len(hyper_1)):
    print (".", end="")
    rf = MLPRegressor(hidden_layer_sizes=(hyper_1[i],  ) )
    hyper_out[i] = evaluate_model_fast(rf, train_x, test_x, train_y,  test_y)
```
